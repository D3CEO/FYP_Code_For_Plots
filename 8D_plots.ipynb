```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

R = 3  
num_bands = 8  
num_phi = 100  
num_theta = 100  

phi = np.linspace(0, np.pi/2, num_phi)  
theta = np.linspace(0, 2*np.pi, num_theta)  
phi, theta = np.meshgrid(phi, theta)

X = R * np.sin(phi) * np.cos(theta)
Y = R * np.sin(phi) * np.sin(theta)
Z = R * np.cos(phi)

band_edges = np.linspace(0, np.pi/2, num_bands + 1)  
band_idx = np.digitize(phi, band_edges) - 1  

cmap = ListedColormap([
    '#e0f7ff',  
    '#b3e0ff',  
    '#80ccff',
    '#4db8ff',
    '#1a99ff',
    '#0077ff',
    '#0062cc',
    '#004999'   
])

facecolors = cmap(band_idx)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=facecolors, linewidth=0, antialiased=False)


ax.set_box_aspect([1, 1, 0.5])  

ax.set_axis_off()


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

R = 3  
num_bands = 8  
num_phi = 100  
num_theta = 100  

phi = np.linspace(0, np.pi/2, num_phi)  
theta = np.linspace(0, 2*np.pi, num_theta)  
phi, theta = np.meshgrid(phi, theta)

X = R * np.sin(phi) * np.cos(theta)
Y = R * np.sin(phi) * np.sin(theta)
Z = R * np.cos(phi)

band_edges = np.linspace(0, np.pi/2, num_bands + 1)  
band_idx = np.digitize(phi, band_edges) - 1  

facecolors = np.ones((*phi.shape, 4))  
equator_band_idx = num_bands - 1 
facecolors[band_idx == equator_band_idx] = [0, 0, 1, 1]  
for i in range(num_bands - 1):  
    facecolors[band_idx == i] = [1, 1, 1, 1]  
border_color = [0.95, 0.95, 0.95, 1]  

for i in range(num_bands - 1):  
    if i != equator_band_idx - 1:  
        facecolors[band_idx == i] = np.array(facecolors[band_idx == i])  
        for j in range(1, len(phi) - 1): 
            for k in range(1, len(theta) - 1): 
                if band_idx[j, k] == i:
                    if band_idx[j - 1, k] != i or band_idx[j + 1, k] != i or band_idx[j, k - 1] != i or band_idx[j, k + 1] != i:
                        facecolors[j, k] = border_color  

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=facecolors, linewidth=0, antialiased=False)
ax.set_box_aspect([1, 1, 0.5])  
ax.set_axis_off()
plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\8D_diagrams\Slushballdiagram.pdf', bbox_inches='tight')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

R = 3  
num_phi = 100  
num_theta = 100  

# Create mesh grid for spherical coordinates
phi = np.linspace(0, np.pi/2, num_phi)  # Latitude from 0 to 90 degrees (hemisphere)
theta = np.linspace(0, 2*np.pi, num_theta)  # Longitude from 0 to 360 degrees
phi, theta = np.meshgrid(phi, theta)

# Convert spherical coordinates to Cartesian coordinates
X = R * np.sin(phi) * np.cos(theta)
Y = R * np.sin(phi) * np.sin(theta)
Z = R * np.cos(phi)

# Fully blue for Unfrozen Earth
r = np.full_like(phi, 0.3)
g = np.full_like(phi, 0.7)
b = np.full_like(phi, 1.0)

colors = np.stack((r, g, b), axis=-1)

# Plot the completely unfrozen hemisphere
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, facecolors=colors, rstride=1, cstride=1, linewidth=0, antialiased=False)

ax.set_box_aspect([1, 1, 0.5])
ax.set_axis_off()
#ax.set_title("Unfrozen Earth (Ice-Free)")
plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\2D_diagrams\unfrozendiagram.pdf', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 3  
num_phi = 100  
num_theta = 100  

# Create mesh grid for spherical coordinates
phi = np.linspace(0, np.pi/2, num_phi)  # Latitude from 0 to 90 degrees (hemisphere)
theta = np.linspace(0, 2*np.pi, num_theta)  # Longitude from 0 to 360 degrees
phi, theta = np.meshgrid(phi, theta)

# Convert spherical coordinates to Cartesian coordinates
X = R * np.sin(phi) * np.cos(theta)
Y = R * np.sin(phi) * np.sin(theta)
Z = R * np.cos(phi)

# Define the latitude split (45 degrees latitude)
boundary = np.pi / 4  

# SWAPPED: Blue at the pole, white at the equator
r = np.where(phi >= boundary, 0.3, 1.0)  
g = np.where(phi >= boundary, 0.7, 1.0)  
b = np.where(phi >= boundary, 1.0, 1.0)  

colors = np.stack((r, g, b), axis=-1)

# Plot the corrected hemisphere
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, facecolors=colors, rstride=1, cstride=1, linewidth=0, antialiased=False)

ax.set_box_aspect([1, 1, 0.5])
ax.set_axis_off()
#ax.set_title("Ice at Equator, No Ice at Pole")
plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\2D_diagrams\moderndiagram.pdf', bbox_inches='tight')
plt.show()
```
