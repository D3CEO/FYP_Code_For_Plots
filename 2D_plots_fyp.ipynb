### LET's Plot these on the Albedo Function

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D 

a1 = 0.7
a2 = 0.3
t1 = 260
t2 = 290
M = 0.1

def a(x):
    return a1 + (a2 - a1) * ((1 + np.tanh(M * (x - ((t1 + t2) / 2)))) / 2)

x = np.linspace(220, 340, 800)
T1b = 243 
T2b = 221
a_T1b = a(T1b)
a_T2b = a(T2b)

T1m = 300
T2m = 228
a_T1m = a(T1m)
a_T2m = a(T2m)

T1t = 326 
T2t = 294
a_T1t = a(T1t)
a_T2t = a(T2t)

plt.figure(figsize=(10, 6))
plt.plot(x, a(x), label=r"$a(T)$", color="blue", linewidth=2)

plt.scatter([T1b, T2b], [a_T1b, a_T2b], color='red', zorder=5)
plt.scatter([T1m, T2m], [a_T1m, a_T2m], color='purple', zorder=5)
plt.scatter([T1t, T2t], [a_T1t, a_T2t], color='green', zorder=5)

plt.text(T1b, a_T1b, r"$T^{1b}$", fontsize=12, ha='left', color='red')
plt.text(T2b, a_T2b, r"$T^{2b}$", fontsize=12, ha='left', color='red')

plt.text(T1m, a_T1m, r"$T^{1m}$", fontsize=12, ha='left', color='purple')
plt.text(T2m, a_T2m, r"$T^{2m}$", fontsize=12, ha='left', color='purple')

plt.text(T1t, a_T1t, r"$T^{1t}$", fontsize=12, ha='left', color='green')
plt.text(T2t, a_T2t, r"$T^{2t}$", fontsize=12, ha='left', color='green')

#plt.title("Plot of $a(T)$ with Typical Points from Stable Branches", fontsize=16)
plt.xlabel("$T$ (K)", fontsize=14)
plt.ylabel(r"$\alpha(T)$", fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  
plt.grid(alpha=0.4)


legend_handles = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=8, label="Points from bottom stable branch"),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='purple', markersize=8, label="Points from middle stable branch"),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=8, label="Points from top stable branch"),
]

plt.legend(handles=legend_handles, fontsize=12, loc='best')

plt.xlim(220, 340)
plt.ylim(0.25, 0.75)

plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\2D_diagrams\2D_Stable_Solutions_on_Albedo_D0p35.pdf', bbox_inches='tight')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

a1 = 0.7
a2 = 0.3
t1 = 260
t2 = 290
M = 0.1

def a(x):
    return a1 + (a2 - a1) * ((1 + np.tanh(M * (x - ((t1 + t2) / 2)))) / 2)

x = np.linspace(250, 300, 500)

T11 = 284.461
T12 = 257.6
T21 = 261.821
T22 = 295.5
a_T11 = a(T11)
a_T21 = a(T21)
a_T12 = a(T12)
a_T22 = a(T22)

plt.figure(figsize=(8, 6))
plt.plot(x, a(x), label=r"$a(x)$", color="blue", linewidth=2)

plt.scatter([T11, T21], [a_T11, a_T21], color="red", label="Bifurcation points", zorder=5)
plt.text(T11, a_T11, f"({T11:.3f}, {a_T11:.3f})", fontsize=10, ha='left', color='red')
plt.text(T21, a_T21, f"({T21:.3f}, {a_T21:.3f})", fontsize=10, ha='left', color='blue')
plt.text(T12, a_T12, f"({T12:.3f}, {a_T12:.3f})", fontsize=10, ha='left', color='green')
plt.text(T22, a_T22, f"({T22:.3f}, {a_T22:.3f})", fontsize=10, ha='left', color='green')
plt.title("Plot of $a(x)$ with Specific Points", fontsize=16)
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$a(x)$", fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.xlim(250,300)
plt.ylim(0.3,0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path_p1 = r"C:\Users\User\Desktop\4th Year\FYP\2D model\Cloudeice\2d_p1vEps.csv"
file_path_p2 = r"C:\Users\User\Desktop\4th Year\FYP\2D model\Cloudeice\2d_p2vEps.csv"

data_p1 = pd.read_csv(file_path_p1)
data_p2 = pd.read_csv(file_path_p2)

segments_p1 = [
    (0, 16, 'blue'),
    (1000, 1011, 'blue'),
    (2000, 2011, 'blue'),
    (3000, 3017, 'blue'),
    (4000, 4014, 'blue'),
    (4087, 4094, 'blue'),
    (5050, 5055, 'blue'),
    (6051, 6057, 'blue'),
    (6133, 6143, 'blue'),
]

segments_p2 = [
    (2000, 2010, 'blue'), 
    (3000, 3016, 'blue'), 
    (6185, 6202, 'blue'), 
    (6685, 6698, 'blue'), 
    (8185, 8192, 'blue'), 
    (8685, 8691, 'blue'), 
    (9185, 9191, 'blue'), 
    (9685, 9693, 'blue'), 
]

fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)

for start, end, color in segments_p1:
    axs[0].plot(data_p1["Epsilon"].iloc[start:end], data_p1["P1"].iloc[start:end], color=color)

axs[0].set_xlabel("$\epsilon$", fontsize=14)
axs[0].set_ylabel("$P_1$", fontsize=14)
axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)
axs[0].grid(True)
axs[0].set_title("Plot of $P_1$ vs $\epsilon$")

for start, end, color in segments_p2:
    axs[1].plot(data_p2["Epsilon"].iloc[start:end], data_p2["P2"].iloc[start:end], color=color)

axs[1].set_xlabel("$\epsilon$", fontsize=14)
axs[1].set_ylabel("$P_2$", fontsize=14)
axs[1].set_xlim(0, 1)
axs[1].set_ylim(0, 1)
axs[1].grid(True)
axs[1].set_title("Plot of $P_2$ vs $\epsilon$")

plt.tight_layout()
plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\2D_diagrams\P1P2vEps.pdf', bbox_inches='tight')
plt.show()
```
