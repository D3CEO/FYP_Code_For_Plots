```python
import numpy as np
import matplotlib.pyplot as plt
Q = 341.3
C = 5e8 
M = 0.1
s = 5.67e-8 
a1 = 0.7
a2 = 0.3
t1 = 260
t2 = 290
eps = 0.8  


def a(x):
    return a1 + (a2 - a1) * ((1 + np.tanh(M * (x - ((t1 + t2) / 2)))) / 2)

def f1(x):
    return (1/C) * (Q * (1 - a(x)))

def f2(x):
    return (1/C) * (eps * s * (x ** 4))

x = np.linspace(200, 320, 500)

intersection_x = 218
intersection_y = f1(intersection_x) 

plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label=r'$\frac{1}{C}Q(1 - a(T))$', color='blue', linewidth=2)
plt.plot(x, f2(x), label=r'$\frac{1}{C} \epsilon \sigma T^4$', color='red', linewidth=2)
plt.scatter(intersection_x, intersection_y, color='black', s=75, zorder=5, label='Stable Fixed Point')
plt.xlabel("Temperature (K)", fontsize=14)
plt.ylabel(r"Rescaled Power Flux (K S$^{-1}$m$^{-2}$)", fontsize=14)
plt.axhline(0, color='black', linewidth=1) 
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\1D_diagrams\Energy_inout1.pdf', bbox_inches='tight')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
Q = 341.3
C = 5e8 
M = 0.1
s = 5.67e-8 
a1 = 0.7
a2 = 0.3
t1 = 260
t2 = 290
eps = 0.5 

def a(x):
    return a1 + (a2 - a1) * ((1 + np.tanh(M * (x - ((t1 + t2) / 2)))) / 2)

def f1(x):
    return (1/C) * (Q * (1 - a(x)))

def f2(x):
    return (1/C) * (eps * s * (x ** 4)) 

x = np.linspace(200, 320, 500)

intersection_x1 = 245.5
intersection_y1 = f1(intersection_x1)

intersection_x2 = 273
intersection_y2 = f1(intersection_x2) 

intersection_x3 = 302.5
intersection_y3 = f1(intersection_x3) 

plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label=r'$\frac{1}{C}Q(1 - a(T))$', color='blue', linewidth=2)
plt.plot(x, f2(x), label=r'$\frac{1}{C} \epsilon \sigma T^4$', color='red', linewidth=2)

plt.scatter(intersection_x1, intersection_y1, color='black', s=90, zorder=5)
plt.scatter(intersection_x2, intersection_y2, color='purple', s=90, zorder=5, label='Unstable Fixed Point')
plt.scatter(intersection_x3, intersection_y3, color='black', s=90, zorder=5, label='Stable Fixed Point')
plt.xlabel("Temperature (K)", fontsize=14)
plt.ylabel(r"Rescaled Power Flux (K S$^{-1}$m$^{-2}$)", fontsize=14)
plt.axhline(0, color='black', linewidth=1) 
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.savefig(r'C:\Users\User\Desktop\4th Year\FYP\Images for Thesis\1D_diagrams\Energy_inout3.pdf', bbox_inches='tight')
plt.show()
```
